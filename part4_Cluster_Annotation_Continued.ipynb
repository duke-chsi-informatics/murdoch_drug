```python
import scanpy as sc
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=80, facecolor='white')
```

# TSNE plot with annotations

```python
combined_tsne = sc.read('/scratch/combined_tsne.h5ad')
combined_umap = sc.read('/scratch/combined_umap.h5ad')
```

```python
# for the annotation to work, the naming of all cell types in tsne-cluster_names must be uniquely presented.
# if there are two clusters with the same type, add empty space after their name to allow for annotations
tsne_cluster_names =list(['NK/endothelial','undetermined','NKT/Mesenchymal stem','Myeloid dendritic/Naive T','NK/NKT','NK/Th17',
                      'dendritic','NKT/T','B','NK','Monocyte/Macrophage','T/NKT','B ','Plasmacytoid dendritic/B'])
```

```python
combined_tsne.rename_categories('louvain', tsne_cluster_names)
sc.pl.tsne(combined_tsne, color='louvain', frameon=False)
sc.pl.tsne(combined_tsne, color=['sample','user_group'], frameon=False)
```

```python
# UMAP plot with annotations
umap_cluster_names =list(['NK/endothelial','undetermined','NKT/Mesenchymal stem','Myeloid dendritic/Naive T','NK/NKT','NK/Th17',
                      'dendritic','NKT/T','B','NK','Monocyte/Macrophage','T/NKT','B ','Plasmacytoid dendritic/B'])
```

```python
combined_umap.rename_categories('louvain', umap_cluster_names)
sc.pl.umap(combined_umap, color='louvain', title='UMAP', frameon=False)
sc.pl.umap(combined_umap, color=['sample','user_group'], frameon=False)
```
